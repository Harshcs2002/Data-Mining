# Use Simple Kmeans, DBScan, Hierachical clustering algorithms for clustering. Compare the performance of clusters by changing the parameters involved in the algorithms.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
from sklearn import datasets
from sklearn.metrics import silhouette_score
```

```python
iris = datasets.load_iris()
iris_data = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_data.head()
```

```python
plt.scatter(iris_data['sepal length (cm)'], iris_data['sepal width (cm)'])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

## KMeans Clustering

```python
estimator1 = KMeans(n_clusters=3)
estimator1.fit(iris_data)
label_pred1 = estimator1.labels_
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(iris_data['petal length (cm)'][label_pred1 == 0], iris_data['petal width (cm)'][label_pred1 == 0], s = 100, c = 'red', label = 'Cluster 0')
plt.scatter(iris_data['petal length (cm)'][label_pred1 == 1], iris_data['petal width (cm)'][label_pred1 == 1], s = 100, c = 'blue', label = 'Cluster 1')
plt.scatter(iris_data['petal length (cm)'][label_pred1 == 2], iris_data['petal width (cm)'][label_pred1 == 2], s = 100, c = 'green', label = 'Cluster 2')

plt.title('KMeans Clustering of Iris Data', fontsize = 20)
plt.xlabel('petal Length (cm)', fontsize = 14)
plt.ylabel('petal width (cm)', fontsize = 14)
plt.legend(fontsize = 12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(iris_data['sepal length (cm)'][label_pred1 == 0], iris_data['sepal width (cm)'][label_pred1 == 0], s = 100, c = 'red', label = 'Cluster 0')
plt.scatter(iris_data['sepal length (cm)'][label_pred1 == 1], iris_data['sepal width (cm)'][label_pred1 == 1], s = 100, c = 'blue', label = 'Cluster 1')
plt.scatter(iris_data['sepal length (cm)'][label_pred1 == 2], iris_data['sepal width (cm)'][label_pred1 == 2], s = 100, c = 'green', label = 'Cluster 2')

plt.title('KMeans Clustering of Iris Data', fontsize = 20)
plt.xlabel('Sepal Length (cm)', fontsize = 14)
plt.ylabel('Sepal Width (cm)', fontsize = 14)
plt.legend(fontsize = 12)
plt.grid(True)
plt.show()
```

```python
silhouette_score(iris_data, label_pred1)
```

```python
SSE = []
for k in range(1, 10):
    estimator = KMeans(n_clusters=k)
    estimator.fit(iris_data)
    SSE.append(estimator.inertia_)
X = range(1, 10)
plt.xlabel('k')
plt.ylabel('SSE')
plt.plot(X, SSE, 'o-')
plt.show()
```
